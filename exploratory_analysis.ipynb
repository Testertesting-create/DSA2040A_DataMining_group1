```python
# Import the pandas library, commonly used for data manipulation and analysis
import pandas as pd

# Define the path to the dataset file (must match the actual filename in your working directory)
file_path = "OSMI_Mental_Health_Final.csv"

# Load the CSV file into a pandas DataFrame.
# This reads the contents of the CSV file into a structured table where rows = records and columns = features.
df = pd.read_csv(file_path)

# Display information about the DataFrame:
# - Number of entries (rows)
# - Number of columns
# - Column names and their data types
# - Count of non-null values per column
df_info = df.info()

# Display the first 5 rows of the DataFrame to get a quick look at the data
df_preview = df.head()

# Output the following for further inspection:
# - The shape of the DataFrame (number of rows and columns)
# - A list of all column names
# - A preview of the first 5 rows
df.shape, df.columns.tolist(), df_preview
```

# 2. Exploratory & Statistical Analysis
### Section 1: Age Distribution

```python
# SECTION 1: AGE DISTRIBUTION
# ----------------------------
# This plot shows how the respondents' ages are distributed.
# Helps identify which age group is most represented.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))  # Set the figure size
sns.histplot(df["Age"], bins=30, kde=True, color="skyblue")  # Histogram with KDE overlay
plt.title("Age Distribution of Respondents")  # Title of the chart
plt.xlabel("Age")  # X-axis label
plt.ylabel("Count")  # Y-axis label
plt.show()  # Display the plot
```

### Detailed Interpretation: Age Distribution

This histogram shows the distribution of ages among survey respondents:

- **Peak Age Range:** Most responses are from individuals aged 20–40, with a peak in the late 20s to early 30s. This suggests the survey is most popular among young adults and early-career professionals.
- **Spread:** The distribution is right-skewed, with fewer older respondents. Very few participants are above 50 years old.
- **Implications:** Conclusions from this dataset may be most applicable to younger adults. The underrepresentation of older adults should be considered when generalizing findings.

Overall, the plot highlights that the survey sample is concentrated among younger adults.

### Section 2: Gender Distribution

```python
# SECTION 2: GENDER DISTRIBUTION
# -------------------------------
# This plot shows how gender is distributed among survey participants.

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Gender", order=df["Gender"].value_counts().index, palette="Set2")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate labels for readability
plt.show()
```

### Detailed Interpretation: Gender Distribution

This bar chart shows the gender breakdown of survey participants:

- **Majority Gender:** The largest group of respondents identifies as male, followed by female. There may also be a small number of non-binary or other gender identities.
- **Diversity:** Multiple gender categories are present, but the distribution is not even.
- **Implications:** Since most responses are from males, analysis and conclusions may reflect male perspectives more strongly. Consider this imbalance when interpreting gender-related results.

Overall, the plot shows some gender diversity, but the sample is predominantly male.

# 9. Crosstab: Treatment vs Work Interference

# 10. Heatmap of Encoded Correlation Matrix

```python
# SECTION 10: CORRELATION HEATMAP
# --------------------------------
# Encode categorical fields for numeric correlation matrix.

from sklearn.preprocessing import LabelEncoder

# Copy and encode relevant columns only
encoded_df = df[['Age', 'Gender', 'Country', 'family_history', 'treatment', 
                 'remote_work', 'benefits', 'care_options', 'wellness_program', 
                 'seek_help', 'leave', 'coworkers', 'supervisor']].copy()

# Encode all string columns using LabelEncoder
label_encoders = {}
for col in encoded_df.columns:
    if encoded_df[col].dtype == 'object':
        le = LabelEncoder()
        encoded_df[col] = le.fit_transform(encoded_df[col])
        label_encoders[col] = le

# Compute correlation matrix
corr_matrix = encoded_df.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Encoded Variables)")
plt.show()
```

### 11. Pairplot for Subset (Age + Treatment Factors)

```python
# SECTION 11: PAIRPLOT – SELECTED FEATURES
# -----------------------------------------
# Use pairplot to explore interaction between variables like age, family history, and treatment.

subset = df[['Age', 'treatment', 'family_history', 'benefits', 'care_options']]
subset_encoded = subset.apply(lambda x: LabelEncoder().fit_transform(x) if x.dtype == 'object' else x)

sns.pairplot(subset_encoded, hue="treatment", palette="husl")
plt.suptitle("Pairwise Relationships Among Selected Features", y=1.02)
plt.show()
```

### Detailed Interpretation: Pairplot of Selected Features

This pairplot explores interactions between age, treatment, family history, benefits, and care options:

- **Variable Relationships:** Each scatterplot shows the relationship between two variables, colored by treatment status.
- **Trends:** Patterns may reveal which combinations of factors are associated with seeking treatment.
- **Implications:** Pairplots help identify clusters and trends that may not be visible in single-variable plots.

Overall, the pairplot provides a comprehensive view of how selected features interact.

### 12. Grouped Summary: Treatment Rates by Age Group

```python
# SECTION 12: TREATMENT RATE BY AGE GROUP
# ----------------------------------------
# Helps show trends across age groups

treatment_by_age_group = df.groupby("age_group")["treatment"].value_counts(normalize=True).unstack()
treatment_by_age_group.plot(kind="bar", stacked=True, figsize=(10,6), colormap="Set3")
plt.title("Treatment Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Proportion")
plt.legend(title="Treatment")
plt.xticks(rotation=0)
plt.show()
```

### Detailed Interpretation: Treatment Rate by Age Group

This bar chart shows the proportion of respondents in each age group who have or have not sought treatment:

- **Age Group Trends:** Some age groups may have higher treatment rates than others.
- **Implications:** Age-related differences in treatment-seeking can inform targeted interventions.
- **Overall:** The plot helps identify which age groups may need more support or outreach for mental health treatment.
