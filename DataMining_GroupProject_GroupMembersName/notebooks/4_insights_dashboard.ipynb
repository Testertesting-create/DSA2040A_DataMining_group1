```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.io as pio
from dash import Dash, dcc, html
import dash_bootstrap_components as dbc

# Load data
df = pd.read_csv('../data/us_model_data/training_model_dataset.csv')

# Prepare age group
df['age_group'] = pd.cut(df['Age'], bins=[10, 25, 40, 60, 100],
                         labels=['Youth', 'Young Adult', 'Adult', 'Senior'])

# Clean up missing values safely
for col in ['self_employed', 'family_history', 'remote_work', 'tech_company',
            'benefits', 'care_options', 'leave', 'treatment', 'age_group']:
    if col in df.columns:
        df[col] = df[col].astype(str).fillna('Unknown')

# Initialize Dash app
app = Dash(__name__, external_stylesheets=[dbc.themes.FLATLY])
server = app.server

# Create Plotly visuals
fig1 = px.histogram(df, x='age_group', color='treatment', barmode='group',
                    title='Treatment Seeking by Age Group', text_auto=True)

fig2 = px.histogram(df, x='self_employed', color='treatment', barmode='group',
                    title='Treatment Seeking by Self-Employment', text_auto=True)

fig3 = px.histogram(df, x='tech_company', color='treatment', barmode='group',
                    title='Treatment Seeking by Tech Industry', text_auto=True)

# Define dashboard layout using Bootstrap
app.layout = dbc.Container(fluid=False, children=[
    html.H1("Mental Health Treatment Dashboard", className="text-center mt-4 mb-4"),

    dbc.Row([
        dbc.Col(dbc.Card([
            dbc.CardHeader("Insight 1"),
            dbc.CardBody([
                html.H5("Treatment Seeking by Age Group", className="card-title"),
                dcc.Graph(figure=fig1)
            ])
        ]), md=12)
    ]),

    dbc.Row([
        dbc.Col(dbc.Card([
            dbc.CardHeader("Insight 2"),
            dbc.CardBody([
                html.H5("Treatment Seeking by Self-Employment", className="card-title"),
                dcc.Graph(figure=fig2)
            ])
        ]), md=6),

        dbc.Col(dbc.Card([
            dbc.CardHeader("Insight 3"),
            dbc.CardBody([
                html.H5("Treatment Seeking by Tech Industry", className="card-title"),
                dcc.Graph(figure=fig3)
            ])
        ]), md=6)
    ])
], style={"padding-bottom": "40px"})

# Run the app
if __name__ == "__main__":
    app.run(debug=True)
```
